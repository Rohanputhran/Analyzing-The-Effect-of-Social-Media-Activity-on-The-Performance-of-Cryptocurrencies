```python
data['content'] = data['content'].apply(lambda x: " ".join(x.lower() for x in x.split()))
data.head()
```

```python
# Data cleaning process:
clean_tweets = []
for tweet in data['content']:
    tweet = re.sub("@[A-Za-z0-9]+","",tweet) #Remove @ sign
    tweet = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", tweet)
    tweet = tweet.replace("#", "").replace("_", " ") #Remove hashtag sign but keep the text
    tweet = ''.join(c for c in tweet if c not in emoji.UNICODE_EMOJI) #Remove Emojis
    clean_tweets.append(tweet)
    

data['Cleantext'] = clean_tweets
df = pd.DataFrame(clean_tweets)
```

```python
data
```

```python
# Removing punctuations
data['Cleantext'] = data['Cleantext'].str.replace('[^\w\s]','')
data
```

```python
# Removing stop words
data['Cleantext']  = data['Cleantext'] .apply(lambda x: " ".join(x for x in x.split() if x not in stop))
data.head()
```

```python
#tokens = nltk.word_tokenize(text)    
 
data['Cleantext_tokenized'] = data.apply(lambda row: nltk.word_tokenize(row['Cleantext']), axis=1)
```

```python
data
```

```python
from nltk.stem import WordNetLemmatizer 
from nltk.corpus import wordnet

wordnet_lemmatizer = WordNetLemmatizer()
def lemmatizer(tweet):
    tweet = [wordnet_lemmatizer.lemmatize(word) for word in tweet]
    return tweet

data['Cleantext_lemmatized'] = data['Cleantext_tokenized'].apply(lambda x: lemmatizer(x))
```
