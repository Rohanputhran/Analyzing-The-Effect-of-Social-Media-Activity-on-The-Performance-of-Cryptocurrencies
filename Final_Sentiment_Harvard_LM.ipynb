```python
# Extracting Bitcoin tweets for 90 days - Each day 500 tweets

Bit_tweets = pd.DataFrame()

for j in range(3):
    MM = 9+j
    for i in range(30):
        start_date = '2021-'+ str(MM) +'-' + str(i+1) 
        end_date = '2021-'+ str(MM) +'-' + str(i+2) 
        str_inp ="Bitcoin + since:" + str(start_date) +' ' "until:" + str(end_date)
        df= pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(str_inp).get_items(), 500))
        if j == 0:
            df["Day"] = i+1
        elif j == 1:
            df["Day"]= 30+i
        else:
            df["Day"] = 60+i
        Bit_tweets = pd.concat([Bit_tweets,df])
    
Bit_tweets.set_index('Day', inplace=True)
Bit_tweets.to_csv('Bit_tweets.csv')

Bit_tweets.head()
```

```python
text_Bit = pd.read_csv("FinalBTCcleandata.csv")['Cleantext_lemmatized']
text_Eth = pd.read_csv("FinalETHcleandata.csv")['Cleantext_lemmatized']
```

```python
# Harvard Sentiment Analysis for Bitcoin

hiv4 = ps.HIV4()
Bit_score_Harvard = []
tokens = []
score_me = []
for i in range(len(text_Bit)):
    text_Bit[i] = str(text_Bit[i])
    tokens = hiv4.tokenize(text_Bit[i])
    score_me = hiv4.get_score(tokens)
    Bit_score_Harvard.append(score_me)

df_Bit_score_Harvard = pd.DataFrame(Bit_score_Harvard)
df_Bit_score_Harvard.columns=["Positive_Har_Bit", "Negative_Har_Bit", "Polarity_Har_Bit", "Subjectivity_Har_Bit"]
df_Bit_score_Harvard
```

```python
# Check for Stationarity by using Augmented Dickey-Fuller test.

def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)
    for key,val in result[4].items():
        out[f'critical value ({key})']=val
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
# Check Stationarity for BTC Harvard Time Series

adf_test(Tmsrs_Bit_Hrvd['Bit_return'])
adf_test(Tmsrs_Bit_Hrvd['Volume'])
adf_test(Tmsrs_Bit_Hrvd['Positive_Har_Bit'])
adf_test(Tmsrs_Bit_Hrvd['Negative_Har_Bit'])
adf_test(Tmsrs_Bit_Hrvd['Polarity_Har_Bit'])
adf_test(Tmsrs_Bit_Hrvd['Subjectivity_Har_Bit'])
```
